```python
# 利用if-elif-else敘述判別輸入字元的種類
ch=input('Input a character:')
if ch.isdigit():			# 是否為數字
    print('A digit')
elif ch.isalpha():		# 是否為英文字母
    print('An alpha')
elif ch.isspace():		# 是否為空格
    print('A space')
else:
    print('Other characters')
```

```python
# 判別輸入的字元，限制只能輸入一個字元
ch=input('Input a character:')
if len(ch)==0 or len(ch)>1:	    # 如果輸入0個或多於1個以上的字元
    print('Input error!')
else:                           # 只輸入一個字元
    if ch.isdigit():
        print('Digit')
    elif ch.isalpha():
        print('A letter with ',end='')
        if ch.isupper():
            print('uppercase')
        else:
            print('lowercase')
    elif ch.isspace():
        print('Space')
    else:
        print('Other character')
```

### 5.5 帶有 else 的迴圈

```python
# 檢查輸入的串列是否有負數（沒有else的寫法）
lst=eval(input('Input a list: '))
found=False  			# 用found來記錄是否有負數被找到
for n in lst:
    if n<0:
        print('Negative value found')
        found=True		# 找到負數
        break
if not found:
    print('No negative value')
```

### 5.6 串列生成式 (List comprehension)

#### 5.6.1 簡單的串列生成式

```python
nums=[3,5,-1,4,9]
```

```python
squ=[]
for i in nums:
    squ.append(i**2)
print(squ)
```

```python
squ=[i**2 for i in nums]
print(squ)
```

```python
squ=[i for i in nums if i>0]
print(squ)
```

```python
nums=[7,9,-8,-2,9,3]
non_z=[]
for i in nums:
    if i<0:
        i=0
    non_z.append(i)
print(non_z)
```

```python
nums=[7,9,-8,-2,9,3]
non_z=[i if i>0 else 0 for i in nums]
print(non_z)
```

```python
nums=[3,6,-6,4,9]
squ=[i**2 if i%2==0 else i for i in nums if i>0]
print(squ)
```

```python
fruits=['apple',
        'kiwi',
        'gava',
        'orange']
```

```python
selection=[]
for fruit in fruits:
    if 'a' in fruit:
        selection.append(fruit)
print(selection)
```

```python
selection=[fruit for fruit in fruits if 'a' in fruit]
print(selection)
```

```python
selection=[fruit.upper() for fruit in fruits if 'a' in fruit]
print(selection)
```
