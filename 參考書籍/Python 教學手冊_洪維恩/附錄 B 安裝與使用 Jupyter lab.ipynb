## 附錄 B 安裝與使用 Jupyter lab 準備工作

### B.1 下載與安裝 Python

<font color=#DFFF00>☘ </font>   Python 的官網： https://www.python.org/

<font color=#DFFF00>☘ </font>  Python 下載鏈接： https://www.python.org/downloads/  (請安裝在 C:\Python3 資料夾內）

<font color=#DFFF00>☘ </font>   測試安裝的 Python
 1. 開啟 cmd 視窗
 2. 切換到 Python 的安裝目錄並啟動 Python：

      c:\Python3\python
 3. 試著鍵入一些 python 的指令看看 （例如鍵入 3+2）

### B.2 建立 Python 的虛擬環境並安裝 Jupyter lab


<font color=#DFFF00>☘ </font>  想建立的虛擬環境 myVenv 在 C:\myWork 裡，請於 cmd 視窗中鍵入下面的指令

> C:\Python3\python.exe -m venv C:\myWork\myVenv  （建立 Python 的虛擬環境）
   
> C:\myWork\myVenv\Scripts\activate.bat    （啟動虛擬環境）
   
> pip install jupyterlab   （安裝JupyterLab）

* JupyterLab 官網： https://jupyter.org/

* [JupyterLab Demo版，可以在瀏覽器裡運行](https://hub.gke2.mybinder.org/user/jupyterlab-jupyterlab-demo-bj5oirac/lab/tree/demo) 


###B.3 啟動與使用 Jupyter lab

<font color=#DFFF00>☘ </font>  要啟動 Jypyter Lab

1. 請先啟動 myVenv。如果現在不在虛擬環境裡，請開啟 cmd 視窗，然後鍵入：

 C:\myWork\myVenv\Scripts\activate.bat

2. 將工作目錄切換到 C:\myWork：

 cd C:\myWork

3. 最後鍵入 jupyter lab 即可啟動 Jupyter lab：

 jupyter lab

<font color=#DFFF00>☘ </font>  如果想為 Jupyter lab 建立捷徑，可將下面的指令打包成 bat 檔

   call C:\myWork\myVenv\Scripts\activate.bat <br>
   cd C:\myWork  <br>
   jupyter lab  <br>
