```python
def clip(lst,min_,max_):
    new_lst=[]
    for n in lst:
        if n<min_:      # 比最小值還小
            new_lst.append(min_)
        elif n>max_:    # 比最大值還大
            new_lst.append(max_)
        else:
            new_lst.append(n)
    return new_lst
```

```python
clip([12,-3,7,300,220],0,255)
```

```python
lst=clip([-2,6,7,40,16],0,10)
```

```python
def remove_negative2(set1):
    set2=set1.copy()
    neg=[i for i in set2 if i<0]
    for e in neg:
        set2.remove(e)
    return set2
```

```python
gi=gen_test()   
```

```python
next(gi)
```

```python
next(gi)
```

```python
gi=gen_test()   
for i in gi:
    print(i)
```

```python
def squ(n):  
    for i in range(1,n+1):
        lst.append(i**2)
    return lst
```

```python
squ(9)   
```

```python
def squ_gen(n):  
    for i in range(1,n+1):
        yield i**2 
```

```python
gi=squ_gen(9)      
print(next(gi))
print(next(gi))
```
