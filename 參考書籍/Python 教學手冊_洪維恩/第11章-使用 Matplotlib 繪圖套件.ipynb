#### 11.1.2 利用 add_subplot() 繪圖

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2,2,32)
y=x**2
```

```python
# 利用 add_subplot() 繪製一個圖形
fig=plt.figure()
ax=fig.add_subplot()   # 沒有給任何參數，代表添加一個子圖
ax.plot(x, y, 'r')
plt.show()
```

```python
# 利用 add_subplot() 同時繪製兩個圖形
x=np.linspace(0,2*np.pi,128)
fig=plt.figure(figsize=(13,4))   # 設定畫布的寬和高 
ax1=fig.add_subplot(1,2,1)       # 建立 1×2 的繪圖區，並取得編號為 1 的繪圖區
ax1.plot(x, np.sin(x), 'r')
ax1.set_title('y=sin(x)')

ax2=fig.add_subplot(1,2,2)       # 取得畫布fig中，編號為 2 的繪圖區
ax2.plot(x, np.cos(x), 'b')
ax2.set_title('y=cos(x)')
plt.show()
```

#### 11.1.3 利用 subplots() 繪圖

```python
# 利用 plt.subplot() 繪製 y=sin⁡(x)*cos⁡(2x) 的圖形
x=np.linspace(0,2*np.pi,128)
fig,ax=plt.subplots()
ax.plot(x, np.sin(x)*np.cos(2*x), 'r')
plt.show()
```

```python
# 利用 plt.subplot() 繪製一個 2×3 的圖形
x=np.linspace(0,2*np.pi,64)
y=np.sin(x)*np.cos(x)**2
fig,ax=plt.subplots(2,3,figsize=(12,6))
ax[0,0].plot(x,y)
ax[0,1].plot(x,y,'ro-')
ax[0,2].plot(x,y,'xk')
ax[1,0].plot(x,y,'g^:')
ax[1,1].plot(x,y,'o')
ax[1,2].plot(x,y,'ms')
plt.show()
```

#### 11.1.4 簡易的繪圖指令

```python
# 利用 plt.plot() 繪製一個3次多項式的圖形
x=np.linspace(-2,5,32)
y=x**3-4*x**2+6
plt.plot(x,y)
plt.title('Cubic poly')
plt.show()
```

```python
# 利用 plt.plot() 同時繪製兩個函數的圖形
x=np.linspace(0,2*np.pi,48)
plt.plot(x,np.sin(x),'r-',x,np.cos(x),'b.')  # 同時繪製兩個函數圖形
plt.xlabel('x-axix')   # x 軸標籤
plt.ylabel('y-axix')   # y 軸標籤
plt.show()
```

#### 11.1.5 儲存繪製的圖形

```python
# 利用 plt.savefig() 儲存繪製的圖形
x=np.linspace(0,6*np.pi,120)
y=np.sin(x)*np.exp(-0.2*x)

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot()
ax.plot(x,y,linewidth=3)     # 指定線條寬度為3個點
ax.grid()     # 加上網格線
plt.savefig("function_plot.png", dpi=300, transparent=True)
plt.show()    # 注意這一行要放在plt.savefig()的後面
```

### 11.2 二維繪圖的修飾

#### 11.2.1 繪圖內容的修飾

```python
# 同時畫兩張圖，並加入圖例(legend)
x=np.linspace(0,2*np.pi,48)
fig,ax=plt.subplots()
ax.plot(x,np.sin(x),label='sin(x)')     # 圖例的標籤為 sin(x)
ax.plot(x,np.cos(x),label='cos(x)',linewidth=5)
ax.legend(loc=3)    # 加入圖例，並置於圖形的左下方
plt.show()
```

```python
# 更改取樣點的數目並加入畫布名
x1=np.linspace(0,2*np.pi,36)
x2=np.linspace(0,2*np.pi,128)

fig,ax=plt.subplots(2,1,figsize=(12,4))
fig.suptitle('Different sampling points',fontsize=16)  # 加入畫布名
ax[0].plot(x1,np.sin(x1**2))
ax[1].plot(x2,np.sin(x2**2),'r')
fig.subplots_adjust(hspace=0.4)  # 調整子圖垂直方向的距離
plt.show()
```

```python
# plot()選項的細部設定，並加入坐標軸名稱
x=np.linspace(0,6,24)
fig,ax=plt.subplots(1,2,figsize=(12,4))
ax[0].plot(x, x*np.sin(x), 'mo-')
ax[1].plot(x, x*np.sin(x),color='blue',marker='s',linestyle='--',
            linewidth=3, markersize=10, markerfacecolor='yellow',
            alpha=0.7)
for i in range(2):     # 利用for迴圈設定坐標軸名稱
    ax[i].set_xlabel('x-axis')
    ax[i].set_ylabel('y-axis')
plt.show()
```

```python
# 設定網格線與坐標軸刻度
x=np.linspace(0,8,32)
y=x*np.sin(x)
fig,ax=plt.subplots(1,2,figsize=(12,4))
ax[0].plot(x,y)
ax[0].grid()  # 設定網格線，或 ax[0].grid(True) 
   
ax[1].plot(x,y,'r')
ax[1].set_yticks([-4,0,4,8])   # 設定坐標軸刻度
ax[1].grid(color='b', alpha=0.5, linestyle=':', linewidth=1)
plt.show()
```

```python
# 在圖形上加上註解文字和箭號
x=np.linspace(0,4*np.pi,200)
fig,ax=plt.subplots()
ax.plot(x,x*np.sin(x),x,x*np.cos(x))
ax.annotate('x sin(x)', xy=(2.5, 2.5), xytext=(2.5, 6), fontsize=12,
            arrowprops=dict(arrowstyle='->',facecolor='black'))
ax.annotate('x cos(x)', xy=(8, -5), xytext=(6, -9),fontsize=12,
            arrowprops=dict(arrowstyle='-',facecolor='black'))
ax.text(4,11,'Annotaion',fontsize=18,color='r')
ax.grid()
plt.show()
```

#### 11.2.2 坐標軸的修飾

```python
# 設定繪圖的範圍
x=np.linspace(0,6,64)
fig,ax=plt.subplots(1,2,figsize=(10,4))
ax[0].plot(x,x**2,x,x**3)   # 繪出兩條曲線
ax[0].set_title('default')

ax[1].plot(x,x**2,'r-', x,x**3,'b--',linewidth=3)
ax[1].set_title('custom range')
ax[1].set_ylim([0,30])  # 設定 y 軸的顯示範圍
fig.suptitle(r'Setting y-limit', fontsize=16)   # 設定畫布標題
plt.show()
```

```python
# 對數坐標
x=np.linspace(0,6,64)
fig,ax=plt.subplots(1,2,figsize=(12,4))
ax[0].plot(x,x**2,x,x**3)
ax[0].set_title('x, log-y')
ax[0].set_yscale('log')    # 設定 y 軸為對數坐標

ax[1].plot(x,x**2,x,x**3)
ax[1].set_title('log-x, log-y')  
ax[1].set_xscale('log')    # 設定 x 軸為對數坐標
ax[1].set_yscale('log')    # 設定 y 軸為對數坐標
plt.show()
```

```python
# 設定刻度和標籤
x=np.linspace(0,6,64)
fig,ax=plt.subplots()
ax.plot(x,np.sin(x),x,np.cos(x))
ax.set_yticks([-1,0,1])     	# 設定 y 軸的刻度
ax.set_xticks([0,3.14,6.28])	# 設定 x 軸的刻度
ax.set_xticklabels([0,r'$\pi$',r'2$\pi$'],fontsize=12)

ax.minorticks_on()      	# 顯示次要刻度
ax.set_yticks([-1,0,1])	# 設定 y 軸的刻度
plt.show()
```

```python
# 以科學記號顯示坐標軸的刻度
x=np.linspace(0,6,64)
fig,ax=plt.subplots(1,2,figsize=(12,4))
ax[0].plot(x,np.exp(2*x))
ax[0].set_title('default')

ax[1].plot(x,np.exp(2*x))
ax[1].set_title('scientific')
ax[1].ticklabel_format(axis='y', style='sci', scilimits=(-3,3))
plt.show()
```

```python
# 設定坐標軸的比例
x=np.linspace(-1,1,200)
y=np.sqrt(1-x**2)
fig,ax=plt.subplots(1,2)
ax[0].plot(x,y,'b',x,-y,'b')
ax[0].set_title('aspect: default')

ax[1].plot(x,y,'b',x,-y,'b')
ax[1].set_aspect('equal')    # 設定坐標軸等比
ax[1].set_title('aspect: equal')
fig.tight_layout()
plt.show()
```

```python
# 設定坐標軸的顯示位置
x=np.linspace(-1.7,6,32)
y=x**3-6*x**2+3*x+2
fig,ax=plt.subplots(1,2,figsize=(12,4))
ax[0].plot(x,y)
ax[0].spines['right'].set_visible(False)    # 設右邊軸線不可見 
ax[0].spines['top'].set_visible(False)      # 設上面軸線不可見 

ax[1].plot(x,y,'r:')
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
ax[1].spines['bottom'].set_position(('data',0)) # 設定下面軸線的位置
ax[1].spines['left'].set_position(('data',0))   # 設定左邊軸線的位置
plt.show()
```

### 11.3 填滿繪圖與極坐標繪圖

#### 11.3.1 填滿兩曲線之間的面積

```python
# 將兩條曲線之間的區域填滿
x=np.linspace(0,2*np.pi,200)
fig,ax=plt.subplots()
ax.plot(x,np.sin(x),'r')
ax.plot(x,np.cos(2*x),'b')
ax.fill_between(x,np.cos(2*x),np.sin(x),alpha=0.5,color='yellow')
plt.show()
```

#### 11.3.2 極坐標繪圖

```python
# 極坐標繪圖
t=np.linspace(0,2*np.pi,64)
fig,ax=plt.subplots(subplot_kw={'projection':'polar'}) # 極坐標系統
ax.plot(t,np.sqrt(t),'r',linewidth=4)
ax.set_rticks([0,1,2])
ax.set_rmax(3)
ax.set_rlabel_position(-45)
plt.show()
```

```python
# 同時繪製極坐標圖與直角坐標圖
t=np.linspace(0,2*np.pi,64)
fig=plt.figure(figsize=(10,4))
ax1=fig.add_subplot(121, projection='polar')   # 極坐標系統
ax1.set_title('polar coordinate')
ax1.set_rticks([-0.5,0,0.5,1])
ax1.plot(t, np.sin(t),'-',linewidth=3)

ax2=fig.add_subplot(122)    # 直角坐標
ax2.plot(t, np.sin(t),'r')
ax2.set_title('x-y coordinate')
ax2.set_aspect(2)
plt.show()
```

```python
# 散佈圖
x=np.linspace(-1,1,30)
y=x+np.random.rand(len(x))
fig,ax=plt.subplots(1,2,figsize=(12,4))
ax[0].scatter(x,y)   # 散佈圖
ax[1].scatter(x,y,marker='^',color='red')   # 修改標記符號和顏色
plt.show()
```

```python
# 細部設定散佈圖資料點的呈現方式
rng=np.random.default_rng(999)
x = rng.random(60)              # x 軸坐標隨機
y = rng.random(60)              # y 軸坐標隨機
colors = rng.random(60)         # 顏色隨機
sizes = 500 * rng.random(60)    # 標記符號大小隨機
fig,ax=plt.subplots()
sc=ax.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar(sc,ax=ax)          # 加入色條
plt.show()
```

#### 11.5.1 等高線圖

```python
# 繪製等高線圖
x = np.linspace(-3, 3, 128)
y = np.linspace(-3, 3, 128)
XX, YY = np.meshgrid(x, y)
ZZ = (1-XX**3+YY**3)*np.exp(-XX**2-YY**2)

fig,ax=plt.subplots(1,2,figsize=(12,4))
cs=ax[0].contour(XX, YY, ZZ, [-0.1,0,0.1,0.3,0.7,0.9])
ax[0].clabel(cs, inline=1, fontsize=8)

ax[1].contour(XX, YY, ZZ, 8, colors='black')
ax[1].contourf(XX, YY, ZZ, 8, alpha=.75, cmap='jet')
plt.show()
```

#### 11.5.2 三維等高線圖與曲面圖

```python
# 三維等高線圖與三維曲面圖
x = np.linspace(-3, 3, 128)
y = np.linspace(-3, 3, 128)
XX, YY = np.meshgrid(x, y)
ZZ = (1-XX**3+YY**3)*np.exp(-XX**2-YY**2)

fig,ax= plt.subplots(1,2,figsize=(9,4),subplot_kw={'projection':'3d'})
colors=['red','blue','black']
ax[0].contour3D(XX,YY,ZZ,[-.1,.1,.7],linewidths=[2,3,4],colors=colors)

ax[1].contour3D(XX,YY,ZZ,[-.1,.1,.7],linewidths=[2,3,4],colors=colors)
ax[1].plot_surface(XX, YY, ZZ, cmap='jet',alpha=0.7)
plt.show()
```

#### 11.5.3 調整三維圖形的仰角和方位角

```python
# 仰角和方位角的設定
x = np.linspace(-10, 10, 36)
y = np.linspace(-10, 10, 36)
XX, YY = np.meshgrid(x, y)
ZZ = np.sin((XX**2+YY**2)**0.5)/(XX**2+YY**2)**0.5

fig,ax= plt.subplots(1,2,figsize=(12,4),subplot_kw={'projection':'3d'})
p=ax[0].plot_surface(XX, YY, ZZ, cmap='hsv')
ax[0].set_yticks([-10,0,10])
ax[0].set_zticks([0,0.5,1])
ax[0].set_title(f'elev={ax[0].elev}, azim={ax[0].azim}')
plt.colorbar(p, ax=ax[0],orientation='vertical', pad=0.05)

ax[1].set_axis_off()
ax[1].plot_surface(XX, YY, ZZ, cmap='jet')
ax[1].view_init(60,-40)
ax[1].set_title(f'elev={ax[1].elev}, azim={ax[1].azim}')
plt.show()
```

#### 11.5.4 三維的散佈圖

```python
# 三維的散佈圖
t = np.linspace(1, 15, 120)
x = np.cos(t)/np.sqrt(t)
y = np.sin(t)/np.sqrt(t)

fig,ax= plt.subplots(1,2,figsize=(10,4),subplot_kw={'projection':'3d'})
p=ax[0].scatter(x, y, t, c=t, cmap='jet')
plt.colorbar(p,ax=ax[0],orientation='vertical',fraction=0.035,pad=0.05)

ax[1].plot(x, y, t,'-')
ax[1].grid(False)
plt.show()
```

### 11.6 動畫的製作

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
# 圓球沿著 y="sin" (2x)  的軌跡移動
x=np.linspace(0,2*np.pi,100)
y=x*np.sin(2*x)
fig, ax =plt.subplots()
dot,=ax.plot([],[],'ro',markersize=12)  # 注意 dot 後面有一個逗號
plt.close()

def init():		# 初始化函數
    ax.plot(x,y)
def animate(i):	# 動畫函數
    dot.set_data(x[i],y[i])

ani=FuncAnimation(fig=fig, func=animate, frames=100,
                    init_func=init, interval=50)
HTML(ani.to_html5_video())
```

```python
# 一起繪出圓球和函數 y=sin(2x) 的軌跡
x=np.linspace(0,2*np.pi,100)
y=x*np.sin(2*x)
fig, ax =plt.subplots()
ax.set_xlim(-0.3,6.5)  # 設定 x 軸範圍
ax.set_ylim(-5.8,5.8)  # 設定 y 軸範圍
dot,=ax.plot([],[],'ro',markersize=12)   # 注意 dot 後面有一個逗號
line,=ax.plot([],[])   # 注意 line 後面有一個逗號
plt.close()

def animate(i):
    dot.set_data(x[i], y[i])
    line.set_data(x[:i], y[:i])

ani=FuncAnimation(fig=fig, func=animate, frames=100, interval=25)
HTML(ani.to_html5_video())
```

```python
# 三維的動畫
x=np.linspace(-6,6,128)
y=np.linspace(-6,6,128)
XX,YY=np.meshgrid(x,y)
ZZ=np.sin(np.sqrt(XX**2+YY**2))
fig,ax= plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(XX,YY,ZZ,cmap='jet',alpha=0.7)
ax.set_axis_off()
plt.close()

def animate(i):
    ax.view_init(60,i)

ani=FuncAnimation(fig=fig,func=animate,frames=np.arange(0,360,3),
                    interval=60)
HTML(ani.to_html5_video())
```
