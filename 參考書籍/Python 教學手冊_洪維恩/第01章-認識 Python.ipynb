<font color=#1E8449>☘ </font>注意程式碼的縮排

```python
# if 敘述區塊
num = 3
if num > 0:
    print('if敘述開始')
    print(num,'是正數') 

print('這行一定會被執行')
```

```python
# 錯誤的if敘述區塊
num = -3 				# 將 3 改為 -3
if num > 0:
    print('if敘述開始')	# 只有這行在 if 區塊內
print(num,'是正數')		# 這行縮排錯誤
    
print('這行一定會被執行')
```

```python
# if區塊內的程式碼未對齊
num = -3                   
if num > 0:
    print('if敘述開始')    
# print(num,'是正數')     # 這行程式碼縮排後沒有對齊
    
print('這行一定會被執行')
```
