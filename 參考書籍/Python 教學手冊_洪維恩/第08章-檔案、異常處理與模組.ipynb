#### 8.2.2 異常處理的語法

```python
try:
    6/0
except ZeroDivisionError:
    print('分母為0')
```

```python
try:
    5>'a'
except TypeError:
    print('型別不符')
```

```python
try:
    'Python'[9]
except IndexError:
    print('索引值不正確')
    
```

```python
# try:
#     'Python'[9]
# except TypeError:        # 沒有捕捉到正確的異常
#     print('索引值不正確')
```

```python
try:
    'Python'[9]
except:                    # 不指明是哪一種異常
    print('索引值不正確')
```

```python
try:
    'Python'[9]
except Exception:
    print('索引值不正確')
```

```python
try:
    'Python'[9]
except Exception as e:
    print(e)
```

```python
try: 
    i+6
except Exception as e:
    print(e)
```

```python
try: 
    a=12+6
except Exception as e:
    print(e)
else:
    print('No exception')
finally:
    print('Done!')
```

```python
# 異常處理的範例
def func(lst,idx,n):
   try: 
      lst[idx]/=n
   except IndexError:            # 索引錯誤
      print('Index out of range')
   except ZeroDivisionError:     # 除數為 0
      print('Divided by zero')
   except Exception as e1:       # 其它錯誤
      print(e1)
   else:
      print(f'No exception,lst={lst}')
   finally:
       print('Exit try-except')
```

```python
func([1,4,7,8],2,10)
```

```python
func([1,4,7,8],2,0)
```

```python
func([1,4,7,8],12,10)
```

```python
func([1,4,7,8],12,0)   
```

```python
func([1,4,7,8],2.0,5)
```

#### 8.2.3 異常處理的應用

```python
# 未經異常處理的程式
while(True):
    n=input('Input a number:')
    if n in 'Qq':
        print('Quit program')
        break
    else:
        print(int(n)**2)
```

```python
# 加入異常處理的程式, 可以濾掉無法轉成 int 的輸入
while(True):
    n=input('Input a number:')
    if n in 'Qq':
        print('Quit program')
        break
    else:
        try: 
            print(int(n)**2)
        except:
            print('Input error')
```
