```python
pa=np.array([[2,4,5],
             [3,3,4]])
```

```python
pb=np.array([[3,4,5,6],
             [0,7,2,1]])
```

```python
s_dist=np.zeros((3,4),dtype=int)   # 定義陣列s_dist
for i in range(3):
    for j in range(4):
        s_dist[i,j]=((pa[:,i]-pb[:,j])**2).sum()  # 計算距離的平方
```

```python
s_dist
```

```python
d=(pa[:,:,None]-pb[:,None,:])
```

```python
d
```

```python
np.sum(d**2,axis=0)
```

```python
def dist_for(pa,pb):
    r,c=pa.shape[1],pb.shape[1]
    dist=np.zeros((r,c),dtype=int)
    for i in range(r):
        for j in range(c):
            dist[i,j]=((pa[:,i]-pb[:,j])**2).sum()
    return dist
```

```python
def dist_brocast(pa,pb):
    return np.sum((pa[:,:,None]-pb[:,None,:])**2,axis=0)
```

```python
%timeit dist_for(pa,pb)
```

```python
%timeit dist_brocast(pa,pb)
```

```python
pa=np.random.randint(0,10,size=(2,400))
pb=np.random.randint(0,10,size=(2,1600))
```

```python
%timeit dist_for(pa,pb)
```

```python
%timeit dist_brocast(pa,pb)
```
